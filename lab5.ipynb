```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import random
```

```python
df = pd.read_csv('/content/Salary_Data.csv')
df.head()
```

```python
x = np.array(df["YearsExperience"])
y = np.array(df["Salary"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
plt.scatter(x_train,y_train)
```

```python
def cost_func (y_cap, y):
    return (1/len(y))*0.5*np.sum((y_cap-y)**2)
```

```python
def step (w,b,x_train,y_train,lr):
    y_cap = x_train * w + b
    dw = (1/len(y_train))*np.sum(np.multiply((y_cap-y_train),x_train))
    db = (1/len(y_train))*np.sum((y_cap-y_train))
    w = w - lr*dw
    b = b - lr*db
    return w,b
```

```python
cost = []
w,b = 100*random.random(), 100*random.random()
for e in range(500):
    w,b = step(w,b,x_train,y_train,1e-3)
    y_cap = x_train * w + b
    cost.append(cost_func(y_cap,y_train))
```
