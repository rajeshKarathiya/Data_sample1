```python
import tensorflow as tf
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
x_train = np.reshape(x_train, (-1,784)).astype(float)
x_test = np.reshape(x_test, (-1,784)).astype(float)
y_train = np.array(y_train).astype(float)
y_test = np.array(y_test).astype(float)
```

```python
i = 76
plt.imshow(np.reshape(x_train[i,:], (28,28)))
y_train[i]
```

```python
class Model(tf.keras.Model):
  def __init__(self,act_fn='relu'):
    super().__init__()
    self.layer1 = tf.keras.layers.Dense(128, activation=act_fn)
    self.layer2 = tf.keras.layers.Dense(64, activation=act_fn)
    self.layer3 = tf.keras.layers.Dense(10, activation='softmax')

  def call(self, x):
    x = self.layer1(x)
    x = self.layer2(x)
    x = self.layer3(x)
    return x
```

```python
model = Model()
loss_fn = tf.keras.losses.BinaryCrossentropy()
```

```python
ip = np.random.rand(1,784)
model(ip).shape
```

```python
model.compile(optimizer="Adam", loss="sparse_categorical_crossentropy", metrics=["acc"])
model.fit(x_train, y_train, epochs=100, batch_size=64)
```

```python
model.evaluate(x_test,y_test)[-1]
```
